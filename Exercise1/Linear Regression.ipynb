```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
ds = pd.read_csv("/home/prasun/Downloads/canada_per_capita_income.csv") 
ds.head()
```

```python
ds.describe()
```

```python
ds.info()
```

```python
# Scatter Plot 
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(ds.year, ds["per capita income (US$)"], c = "#175382", marker= ".")
```

```python
lin_reg = linear_model.LinearRegression()
lin_reg.fit(ds[["year"]],ds["per capita income (US$)"])
```

```python
print(f"Coefficient: {lin_reg.coef_}")
print(f"Intercept: {lin_reg.intercept_}")
```

 **Linear Equation became**
 `y = (828.46507522)x + (-1632210.7578554575)`

```python
pr = pd.DataFrame(data={'year': [2016,2020]})
pr
```

```python
lin_reg.predict(pr[["year"]])
```
