```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
dfs= pd.DataFrame(data=scale.fit_transform(df), columns= ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'])
dfs.head()
```

```python
# Before feature Scaling
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize= (10,8))
plt.boxplot(df)
```

```python
# After Feature Scaling
plt.figure(figsize= (10,8))
plt.boxplot(dfs)
```
