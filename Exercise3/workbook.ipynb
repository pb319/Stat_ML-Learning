```python
#Let's make a dataset
# y = 3x + 2
x = np.array([1,2,3,4,5,6])
y = (x*3) + 2
```

```python
import math
# The Gradient Descent[w=3, b=2]
def Grades(x,y, iter = 10000,l = 0.1):
    m = len(x)
    w = b = 1
    cost = 0
    for i in range(iter):
        yp = (w*x)+b
        W = (1/m)*np.sum((yp-y)*x)
        B = (1/m)*np.sum(yp-y)
        w = w - (l*W)
        b = b - (l*B)
        if(math.isclose(cost,(1/(2*m))*np.sum((yp-y)**2),abs_tol=1e-20) and i != 0):
            print(f"FINAL: w: {w}, b: {b}, cost = {(1/(2*m))*np.sum((yp-y)**2)}, Iteration: {i}")
            break;

        print(f"w: {w}, b: {b}, cost = {(1/(2*m))*np.sum((yp-y)**2)}, Iteration: {i}")
        # if(i in [1,10,100,20,500,600,750,900,990,1000]):
        #     print(f"w: {w}, b: {b}, cost = {(1/(2*m))*np.sum((yp-y)**2)}, Iteration: {i}")
        

Grades(x,y)
    
```
