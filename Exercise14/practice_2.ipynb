```python
from sklearn.preprocessing import LabelEncoder
lbn = LabelEncoder()
df["Spam"] = lbn.fit_transform(df["Category"]) 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cu = CountVectorizer()
X_train_count = cu.fit_transform(df["Message"])
y_train_count  = df.Spam
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
import numpy as np
np.mean(cross_val_score(MultinomialNB(),X_train_count,y_train_count))
```

```python
# Pipeline Formation
from sklearn.pipeline import Pipeline
clf = Pipeline([("Vectorizer", CountVectorizer()),
                ("NB",MultinomialNB())])
```

```python
see= clf.fit(df.Message, df.Spam)
```

```python
clf.score(df.Message, df.Spam)
```

```python
df.Category.apply(lambda x: 1 if x == "spam" else 0)
from sklearn.model_selection import StratifiedKFold
sk = StratifiedKFold(n_splits= 10)
sc = []
for (train,test) in sk.split(df.Message,df.Category):
    X_train, y_train, X_test, y_test = df.Message.iloc[train],df.Category.iloc[train],df.Message.iloc[test],df.Category.iloc[test]
    final.fit(X_train,y_train)
    sc.append(final.score(X_test,y_test))
import numpy as np
print(f"Score: {np.mean(sc)}")
```
