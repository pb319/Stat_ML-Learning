```python
import pandas as pd
```

```python
df = pd.read_csv("/home/prasun/GitDemo/Stat_ML-Learning/Exercise5/data.csv")
df.head()
```

```python
dumm = pd.get_dummies(df["Car Model"],dtype= 'int', drop_first= True) 
```

```python
ddf = pd.concat([df,dumm], axis= "columns")
ddff = ddf.drop("Car Model", axis= "columns")
```

```python
ddff.head()
```

```python
from pandas.plotting import scatter_matrix
atb = ['Mileage','Sell Price($)','Age(yrs)','BMW X5','Mercedez Benz C class']
scatter_matrix(ddff[atb],figsize=(20,16))
```

```python
# Instantiating object 
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
X= ddff.drop(["Sell Price($)"],axis="columns")
y = ddff["Sell Price($)"]
reg.fit(X,y)
```
