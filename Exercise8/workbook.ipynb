```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
# Have a look 
iris.data[0]
```

```python
# Train_Test_Slpit
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size= 0.25, random_state= 1)
```

```python
# Model Fitting
from sklearn.linear_model import LogisticRegression
leg = LogisticRegression()
leg.fit(X_train,y_train)
```

```python
leg.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
com = confusion_matrix(y_test,leg.predict(X_test))
com
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(7,5))
sns.heatmap(com,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
