```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
# Lets see What it Contains
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:4]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
leg = LogisticRegression()
```

```python
leg.fit(X_train,y_train)
```

```python
leg.score(X_test,y_test) 
```

```python
plt.matshow(digits.images[91])
```

```python
digits.target[91]
```

```python
leg.predict([digits.data[91]])
```

```python
leg.predict(digits.data[5:15])
```

```python
digits.target[5:15]
```

```python
# Confusion Matrics
from sklearn.metrics import confusion_matrix
pred = leg.predict(X_test)
com = confusion_matrix(y_test,pred)
com
```

```python
# Lets Visualize Confusion Matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
sns.heatmap(com,annot= True)
plt.ylabel("Truth")
plt.xlabel("Predicted")
```
