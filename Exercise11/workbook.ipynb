```python
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
import pandas as pd
df = pd.DataFrame(data= iris.data, columns= iris.feature_names)
df.head() 
```

```python
df["target"] = iris.target
df.head()
```

```python
X = df.drop(["target"],axis = "columns")
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size= 0.2, random_state= 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
raf1 = RandomForestClassifier(criterion= "log_loss")
raf2 = RandomForestClassifier(n_estimators= 10)
raf1.fit(X_train,y_train)
raf2.fit(X_train,y_train)
```

```python
raf1.score(X_test,y_test)
```

```python
raf2.score(X_test,y_test)
```

```python
pred1 = raf1.predict(X_test)
pred2 = raf2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test,pred1)
cm2 = confusion_matrix(y_test,pred2)
```
