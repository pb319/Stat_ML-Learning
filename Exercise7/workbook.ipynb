```python
d1 = df[["satisfaction_level","number_project","time_spend_company","Work_accident"]]
```

```python
dum = pd.get_dummies(df['salary'], dtype= int, drop_first= True)
```

```python
df2  = pd.concat([d1,dum],axis="columns")
df2.head()
```

```python
sns.countplot(hue="salary", x="left", data=df)
```

```python
sns.countplot(hue="Department", x="left", data=df)
```

```python
df2.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(df2,df[["left"]],test_size= 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
leg = LogisticRegression()
leg.fit(X_train,y_train.values.ravel())
pred = leg.predict(X_test)
```

```python
leg.score(X_test,y_test) #80.43% accuracy
```
